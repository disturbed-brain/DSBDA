```python
import nltk
import math
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
```

```python
documentA = 'Jupiter is the largest Planet'
documentB = 'Mars is the fourth planet from the Sun'
```

```python
documentA = documentA.lower()
documentB = documentB.lower()
```

```python
nltk.download('punkt_tab')
wordsA = word_tokenize(documentA)
wordsB = word_tokenize(documentB)
```

```python
stop_words = set(stopwords.words('english'))

filteredA = [w for w in wordsA if w not in stop_words]
filteredB = [w for w in wordsB if w not in stop_words]
```

```python
uniqueWords = set(filteredA).union(set(filteredB))
```

```python
numOfWordsA = dict.fromkeys(uniqueWords, 0)
numOfWordsB = dict.fromkeys(uniqueWords, 0)

for word in filteredA:
    numOfWordsA[word] += 1

for word in filteredB:
    numOfWordsB[word] += 1
```

```python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    totalWords = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(totalWords)
    return tfDict

tfA = computeTF(numOfWordsA, filteredA)
tfB = computeTF(numOfWordsB, filteredB)

print("\nTF for Document A:", tfA)
print("TF for Document B:", tfB)
```

```python
def computeIDF(documents):
    N = len(documents)
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict

idfs = computeIDF([numOfWordsA, numOfWordsB])
print("\nIDF:", idfs)
```

```python
def computeTFIDF(tf, idfs):
    tfidf = {}
    for word, val in tf.items():
        tfidf[word] = val * idfs[word]
    return tfidf

tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)

print("\nTF-IDF for Document A:", tfidfA)
print("TF-IDF for Document B:", tfidfB)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('wordnet')
```

```python
text = "studies studying cries cry"
words = word_tokenize(text)
print("Original Words:", words)
```

```python
ps = PorterStemmer()
stemmed_words = []

for word in words:
    stemmed_words.append(ps.stem(word))

print("\nStemmed Words:", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = []

for word in words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

print("\nLemmatized Words:", lemmatized_words)
```
